```python
df1 = df.copy()
```

```python
df1.drop(["add_to_cart_order", "order_number", "order_dow", "order_hour_of_day", "aisle_id", "department_id", "department", "aisle"], inplace= True, axis= 1)
```

```python
df1.head()
```

```python
banana_order = df1[df1.product_id == 13176].order_id.unique().tolist()
user = df1.user_id.unique().tolist()
```

```python
df_banana_user = df1[df1.user_id.isin(user)]
```

```python
df_banana_user["banana_order"] = 0
df_banana_user.loc[df_banana_user.order_id.isin(banana_order), "banana_order"] = 1
```

```python
df_banana_user.groupby(["user_id", "order_id"], as_index= False).agg(days_since_prior_order= ("days_since_prior_order", "mean"), banana_order= ("banana_order", "mean"))
```

```python
df2 = df_banana_user.groupby(["user_id", "order_id"], as_index= False).agg(days_since_prior_order= ("days_since_prior_order", "mean"), banana_order= ("banana_order", "max")).fillna(0).sort_values(by= ["user_id", "order_id"], ascending= True)
```

```python
import numpy as np
df2["days"] = np.where(df2.banana_order == 0, df2["days_since_prior_order"].sum(axis=0), 0)
```
